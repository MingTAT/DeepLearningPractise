```python
import pandas as pd
import numpy as np
from numpy import *
from random import *
import matplotlib.pyplot as plt
```

```python
def Matr(x, y, order):
    X = []
    for i in range(order + 1):
        X.append(x**i)
    X = mat(X).T
    Y = array(y).reshape((len(y),1))
    return X, Y
```

```python
def fig(x1, y1):
    plt.xlabel('Speed')
    plt.ylabel('Distance')
    #plt.plot(x, y, 'o')
    plt.plot(x1, y1, color = 'g', linestyle = '-', marker = '')
    plt.plot(x1, y1, '*')
    plt.grid()
    plt.show()
```

```python
def order(order):
    df = pd.read_csv('cars.csv')
    x = df['speed']
    y = df['dist']
    plt.plot(x, y, 'o')
    plt.show()
    X, Y = Matr(x, y, order)
    XT = X.transpose()
    B = dot(dot(linalg.inv(dot(XT, X)), XT), Y)
    myY = dot(X, B)
    fig(x, myY)
    
    #apply least square to get the samllest Error Estimate
```
